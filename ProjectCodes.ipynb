## Importing Libraries

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import nltk
nltk.download('vader_lexicon')
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df= pd.read_csv("/content/2021_12_AllnLatestETNewsLastArchive.csv")
```

```python
df.head()
```

```python
sent_analyser = SentimentIntensityAnalyzer()
def sentiment(review):
    return (sent_analyser.polarity_scores(review)["compound"])
```

```python
df['Polarity']= df['Headline'].apply(sentiment)
```

```python
df.head()
```

```python
data1=df.drop(['Unnamed: 0','Content'],axis=1)
```

```python
def senti(data1):
    if( data1['Polarity'] >0 and data1['Polarity']<1):
        val = "Positive"
    elif (data1['Polarity'] <0 and data1['Polarity']>-1):
        val = "Negative"
    else:
        val = "Neutral"
    return val
```

```python
data1['Sentiment'] = data1.apply(senti, axis=1)
```

```python
data1
```

```python
data1.info()
```

```python
data1["Date"]= pd.to_datetime(data1["Date"])
```

```python
data1
```

```python
data1.info()
```

```python
data1 = data1.groupby([data1['Date'].dt.date]).mean()
```
